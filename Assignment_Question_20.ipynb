```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def find_maximum_subtree_sum(root):
    if root is None:
        return 0
    
    # Recursive calls to calculate the subtree sums of the left and right subtrees
    left_sum = find_maximum_subtree_sum(root.left)
    right_sum = find_maximum_subtree_sum(root.right)
    
    # Calculate the current subtree sum
    current_sum = left_sum + right_sum + root.data
    
    # Update the maximum sum if the current sum is greater
    if current_sum > find_maximum_subtree_sum.max_sum:
        find_maximum_subtree_sum.max_sum = current_sum
    
    # Return the sum of the current subtree
    return current_sum

def find_maximum_sum_subtree(root):
    if root is None:
        return 0
    
    # Initialize the maximum sum as negative infinity
    find_maximum_subtree_sum.max_sum = float('-inf')
    
    # Call the recursive function to find the maximum sum
    find_maximum_subtree_sum(root)
    
    return find_maximum_subtree_sum.max_sum

# Example usage
# Creating the binary tree from the example input
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Finding the subtree with maximum sum
maximum_sum = find_maximum_sum_subtree(root)

print("Maximum subtree sum:", maximum_sum)
```

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

        
def construct_bst(level_order):
    if not level_order:
        return None

    root = Node(level_order[0])
    queue = [root]
    i = 1

    while i < len(level_order):
        current = queue[0]
        queue = queue[1:]

        if i < len(level_order) and level_order[i] < current.data:
            current.left = Node(level_order[i])
            queue.append(current.left)
            i += 1

        if i < len(level_order) and level_order[i] > current.data:
            current.right = Node(level_order[i])
            queue.append(current.right)
            i += 1

    return root
# Example usage
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = construct_bst(level_order)
        
        
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

</aside>

```python
def is_valid_level_order(arr, n):
    if n == 0:
        return True

    i = 1
    while i < n:
        if arr[i] < arr[0]:
            i += 1
        else:
            break

    for j in range(i, n):
        if arr[j] < arr[0]:
            return False

    left = is_valid_level_order(arr[1:i], i-1)
    right = is_valid_level_order(arr[i:], n-i)

    return left and right
# Example usage
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
result1 = is_valid_level_order(arr1, len(arr1))
print("Output1:", "Yes" if result1 else "No")

arr2 = [11, 6, 13, 5, 12, 10]
result2 = is_valid_level_order(arr2, len(arr2))
print("Output2:", "Yes" if result2 else "No")
```
